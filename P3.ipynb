```python
import pandas as pd
import numpy as np
df=pd.DataFrame({"enroll":range(240401,240411)})
df["Date"]=pd.to_datetime(np.random.choice(pd.date_range("2020-01-01","2020-12-31"),size=len(df)))
df["Gender"]=np.random.choice(("Male","Female"),size=10,p=[0.9,0.1])
df["Job"]=np.random.choice(("Finance","Accountante","Job Less"),size=10,p=[0.3,0.3,0.4])
df["sal"]=np.random.randint(1000,2000, size=10)
df.loc[df["Job"]=='Job Less',"sal"]=0
df["Day"] = df["Date"].dt.day_name()
display(df)
```

```python
d=df.groupby("Job")["sal"].sum()
d
e=df.groupby("Job")["sal"].mean()
e
r=df.groupby("Job")["sal"].std()
r
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.DataFrame({"Serial no.":range(1,101)})
df["Serial no."]=pd.to_datetime(np.random.choice(pd.date_range("2020-01-31","2020-12-31"),size=len(df)))
df["Gender"]=np.random.choice(("Male","Female"),size=(100),p=[0.5,0.5])
df["Department"]=np.random.choice(("a","f","hr","it"),size=100,p=[0.25,0.25,0.25,0.25])
df["job type"]=np.random.choice(("Part time","full time"),size=100,p=[0.2,0.8])
df["Rating"]=np.random.choice((1,2,3,4,5,6,7,8,9),size=100)

df["Month"]=df["Serial no."].dt.month
df
```

```python
d=df.groupby(["Month","Department"]).size().unstack(fill_value=0)
plt.subplot(1,2,1)
plt.figure(figsize=(12,5))
d.plot(kind="bar")
plt.subplot(1,2,2)
d.plot(kind="line")
```

```python
import pandas as pd
import numpy as np
df=pd.DataFrame({"id":range(1,11)})
df["d"]=pd.to_datetime(np.random.choice(pd.date_range("2020-01-01","2020-12-31"),size=len(df)))
df["Sal"]=np.random.randint(1000,10000,size=10)
df["dep"]=np.random.choice(("hr","it","acc"),size=10,p=[0.3,0.3,0.4])
df["month"]=df["d"].dt.month
df
```

```python
d=df.groupby(["month","dep"]).size().unstack(fill_value=0)
d.plot(kind="bar")
```
