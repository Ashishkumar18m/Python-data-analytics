```python
import numpy as np
from scipy.stats import t

ci_data = []

for year, group in rice_df.groupby('Year'):
    data = group['RICE PRODUCTION (1000 tons)_mean']
    n = len(data)
    mean = data.mean()
    std = data.std()
    
    t_value = t.ppf(0.975, df=n-1)
    margin = t_value * (std / np.sqrt(n))
    
    lower = mean - margin
    upper = mean + margin
    
    ci_data.append([year, mean, lower, upper])

ci_df = pd.DataFrame(ci_data, columns=['Year', 'Mean', 'CI Lower', 'CI Upper'])
print(ci_df)

```

### <span style='color :blue'> 4.4 Type I and Type II errors and their relevance in your context.</span>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Create x-axis values
x = np.linspace(-4, 8, 500)

# Define distributions
mu0, sigma0 = 0, 1               # Null hypothesis distribution
mu1, sigma1 = 2, 1               # Alternative hypothesis distribution

# PDFs
y0 = norm.pdf(x, mu0, sigma0)
y1 = norm.pdf(x, mu1, sigma1)

# Significance level α (5% right-tailed test)
alpha = 0.05
critical_value = norm.ppf(1 - alpha, mu0, sigma0)

plt.figure(figsize=(12, 6))

# Plot both distributions
plt.plot(x, y0, label="H0: No Change in Mean Rice Production", linewidth=2)
plt.plot(x, y1, label="H1: Mean Rice Production Changed", linewidth=2)

# Shade Type I error (α)
x_alpha = np.linspace(critical_value, 8, 300)
plt.fill_between(x_alpha, norm.pdf(x_alpha, mu0, sigma0), color="red", alpha=0.3, label="Type I Error (α)")

# Shade Type II error (β)
x_beta = np.linspace(-4, critical_value, 300)
plt.fill_between(x_beta, norm.pdf(x_beta, mu1, sigma1), color="blue", alpha=0.3, label="Type II Error (β)")

# Critical line
plt.axvline(critical_value, color='black', linestyle='--', linewidth=2)
plt.text(critical_value + 0.1, 0.1, "Critical Value", rotation=90)

plt.title("Type I and Type II Errors in Rice Production Hypothesis Test")
plt.xlabel("Test Statistic")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```

```python

import pandas as pd
from pathlib import Path

path = "Year_State_Crop_Statistics.csv"   # change path if different
assert Path(path).exists(), f"{path} not found"

# read with multi-row header (your file uses multi-level headers)
df = pd.read_csv(path, header=[0,1])

# flatten multiindex: join non-empty levels with underscore, remove trailing underscores
df.columns = [
    "_".join([str(a).strip() for a in col if str(a).strip() != ""]).replace(" ", " ")
    for col in df.columns.values
]

# Quick rename of the left-most columns if they were unnamed
# (based on your earlier output)
df = df.rename(columns={
    'Unnamed: 0_level_0_Unnamed: 0_level_1': 'Year',
    'Unnamed: 1_level_0_Unnamed: 1_level_1': 'State'
})

# Show shape and first rows to verify
print("Data shape:", df.shape)
display(df.head(3))
print("\nColumns sample:", df.columns.tolist()[:30])
```

```python
# target name (edit if different)
target_col = "RICE PRODUCTION (1000 tons)_mean"

# verify target exists
if target_col not in df.columns:
    raise KeyError(f"Target column {target_col} not found. Available cols with 'rice':\n{[c for c in df.columns if 'rice' in c.lower()]}")

# create dataset with numeric features
drop_cols = ['Year', 'State', target_col]
feature_cols = df.select_dtypes(include=['int64','float64']).columns.difference(drop_cols).tolist()

# optionally remove columns that look like skew/kurtosis if you prefer (keep means, yields, area)
# e.g., keep only columns containing 'mean' or 'YIELD' or 'AREA' or 'PRODUCTION' as heuristics
feature_cols = [c for c in feature_cols if ('mean' in c.lower() or 'yield' in c.lower() or 'area' in c.lower() or 'production' in c.lower())]

print("Number of features:", len(feature_cols))
print("Some features:", feature_cols[:20])

# build modeling dataframe
model_df = df[['Year','State', target_col] + feature_cols].copy()

# drop rows with missing target, drop rows with too many NaNs
model_df = model_df.dropna(subset=[target_col])
model_df = model_df.dropna(thresh=int(0.7 * len(model_df.columns)))  # require 70% present

print("Modeling rows after basic cleaning:", model_df.shape[0])
display(model_df.head(3))
```

```python
from sklearn.model_selection import train_test_split

X = model_df[feature_cols].fillna(0)   # simple imputation: 0 for missing numeric features (you can replace with median)
y = model_df[target_col]

# recommended: stratify if you want balanced years, but for regression we just shuffle
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

print("Train shape:", X_train.shape, "Test shape:", X_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

# pipeline: scale features then linear regression
pipeline = make_pipeline(StandardScaler(), LinearRegression())
pipeline.fit(X_train, y_train)

# predictions
y_pred = pipeline.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import numpy as np

r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae = mean_absolute_error(y_test, y_pred)

print("Linear Regression performance on test set")
print(f"R^2: {r2:.4f}")
print(f"RMSE: {rmse:.3f}")
print(f"MAE: {mae:.3f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

residuals = y_test - y_pred

plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
sns.scatterplot(x=y_pred, y=residuals)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel("Predicted")
plt.ylabel("Residual")
plt.title("Residuals vs Predicted")

plt.subplot(1,2,2)
sns.histplot(residuals, kde=True)
plt.title("Residual distribution")
plt.xlabel("Residual")
plt.show()
```

```python
from sklearn.model_selection import cross_val_score, KFold
import numpy as np

kf = KFold(n_splits=5, shuffle=True, random_state=42)
cv_r2 = cross_val_score(pipeline, X, y, cv=kf, scoring='r2')
cv_rmse = np.sqrt(-cross_val_score(pipeline, X, y, cv=kf, scoring='neg_mean_squared_error'))
cv_mae = -cross_val_score(pipeline, X, y, cv=kf, scoring='neg_mean_absolute_error')

print("Cross-val (5-fold) R2 mean:", np.mean(cv_r2).round(4), "std:", np.std(cv_r2).round(4))
print("Cross-val RMSE mean:", np.mean(cv_rmse).round(3))
print("Cross-val MAE mean:", np.mean(cv_mae).round(3))
```

#### Elbow Method Plot (For Choosing K)

```python
# Compute Distortions (Elbow Method)
from sklearn.cluster import KMeans

distortions = []

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    distortions.append(kmeans.inertia_)
```

```python
# Plot the Elbow Curve
plt.figure(figsize=(6,4))
plt.plot(range(1, 11), distortions, marker='o')
plt.title("Elbow Method for Optimal K")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Distortion (Inertia)")
plt.grid(True)
plt.show()
```
