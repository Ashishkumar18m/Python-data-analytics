## merge 2 inventory

```python
import pandas as pd
invt1={
    'pen':[2],
    'apple':[1],
    'pencil':[2],
    'notebook':[4]
}
invt2={
    'pen':[4],
    'orange':[2],
    'apple':[3],
    'notebook':[2]
}

df=pd.DataFrame(invt1)
df1=pd.DataFrame(invt2)
print("invt 1\n:",df)
print("invt 2\n:",df1)
d=df.add(df1,fill_value=0)
print(d)
df_long = pd.concat([df, df1], keys=["invt1", "invt2"]).stack()
print(df_long)
df_T = df.T
df1_T = df1.T
inventory = pd.merge(df_T, df1_T, left_index=True, right_index=True, how='outer', suffixes=('_invt1', '_invt2'))
print(inventory)

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create figure
fig, ax = plt.subplots(figsize=(8, 5), dpi=120)

# Plot line with custom style
ax.plot(
    x, y,
    color='crimson',       # line color
    linewidth=2.5,         # line width
    linestyle='-',         # solid line
    marker='o',            # marker shape
    markersize=6,          # marker size
    markerfacecolor='white',
    markeredgecolor='black',
    label='Sine Wave'
)

# Title & Labels with font styling
ax.set_title(
    "Beautiful Matplotlib Graph",
    fontsize=18,
    fontweight='bold',
    color='navy',
    pad=15
)
ax.set_xlabel("X Axis", fontsize=14, labelpad=10, color='darkgreen')
ax.set_ylabel("Y Axis", fontsize=14, labelpad=10, color='darkgreen')

# Grid style
ax.grid(
    True,
    which='major',
    linestyle='--',
    linewidth=0.5,
    color='gray',
    alpha=0.7
)
ax.minorticks_on()
ax.grid(
    True,
    which='minor',
    linestyle=':',
    linewidth=0.3,
    color='gray',
    alpha=0.5
)

# Axis limits
ax.set_xlim(0, 10)
ax.set_ylim(-1.2, 1.2)

# Background colors
ax.set_facecolor('#fef9e7')  # plot background
fig.patch.set_facecolor('#e8f8f5')  # outer background

# Spines (borders) customization
ax.spines['top'].set_visible(False)   # show/hide top spine
ax.spines['right'].set_visible(False) # show/hide right spine
ax.spines['bottom'].set_linewidth(2)
ax.spines['bottom'].set_color('black')
ax.spines['left'].set_linewidth(2)
ax.spines['left'].set_color('black')

# Removing boundaries example:
# Uncomment to remove all spines
# for spine in ax.spines.values():
#     spine.set_visible(False)

# Ticks customization
ax.tick_params(axis='x', colors='purple', direction='in', length=6, width=1.5)
ax.tick_params(axis='y', colors='purple', direction='in', length=6, width=1.5)

# Legend styling
legend = ax.legend(
    loc='upper right',
    fontsize=12,
    frameon=True,
    facecolor='white',
    edgecolor='black'
)

# Text annotation
ax.text(
    7, 0.5, 'Peak here!',
    fontsize=12,
    color='darkred',
    bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3')
)

# Save the figure with high quality
plt.savefig('beautiful_plot.png', dpi=300, bbox_inches='tight')

# Show the plot
plt.show()
```
