```python
import pandas as pd
#import numpy as np

# First DataFrame
data1 = {
    'product': ['apple', 'banana', 'orange', 'grape'],
    'price': [50, 30, 40, 60],
    'quantity': [100, 150, 80, 60]
}
df1 = pd.DataFrame(data1)
print("First Data:\n",df1)

# Second DataFrame
data2 = {
    'product': ['apple', 'banana', 'kiwi', 'grape'],
    'price': [55, 35, 45, 60],
    'quantity': [90, 130, 50, 65]
}
df2 = pd.DataFrame(data2)
print("Second Data:\n",df2)
```

```python
# Add quantities (combine inventories):
df_combined = df1.add(df2, fill_value=0)
print(df_combined)
```

```python
df_merge=pd.merge(df1, df2, on='product') 
print(df_merge)
```

```python
df1.compare(df2)
```

```python
df1.equals(df2)
```

```python
pd.concat([df1, df2])
```

```python
df1.set_index('product').combine_first(df2.set_index('product'))
```

# Same Data Except 1 --  concat,add,subtract

```python
import pandas as pd
invt1={
    'pen':[2],
    'apple':[1],
    'pencil':[2],
    'notebook':[4]
}
invt2={
    'pen':[4],
    'orange':[2],
    'apple':[3],
    'notebook':[5]
}

df3=pd.DataFrame(invt1)
df4=pd.DataFrame(invt2)
print("invt 1\n:",df3)
print("invt 2\n:",df4)
```

```python
# Concatenate DataFrames
pd.concat([df3, df4])
```

```python
pd.merge(df3, df4, on="pen")
```

```python
# Add values (like +).
df3.add(df4, fill_value=0)
```

```python
# 	Subtract values
df4.sub(df3, fill_value=0)
```

```python
df3.mul(df4)
```

```python
# Check if they're the same
df3.equals(df4)
```

# Same data - Product   --  == , != ,  > < <= >= , .compare	, .equals

```python
import pandas as pd
invt1={
    'pen':[2],
    'apple':[1],
    'pencil':[2],
    'notebook':[4]
}
invt2={
    'pen':[4],
    'apple':[3],
    'pencil':[2],
    'notebook':[5]
}

df5=pd.DataFrame(invt1)
df6=pd.DataFrame(invt2)
print("invt 1\n:",df5)
print("invt 2\n:",df6)
```

```python
# 	Check where values are equal
df5 == df6
```

```python
# Check where values differ.
df5!= df6
```

```python
# Element-wise comparison.
#df5 > df6
df5 >=df6
#df5 < df6
#df5 <= df6
```

```python
# Shows differences between two DataFrames.
df5.compare(df6)
```

```python
# Returns True if DataFrames are exactly the same.
df5.equals(df6)
```
