```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.DataFrame({"id":range(1,11)})
df["Joining data"]=pd.to_datetime(np.random.choice(pd.date_range("2020-01-01","2020-12-31"),size=len(df)))
df["Gender"]=np.random.choice(("Male","Female"),size=10,p=[0.5,0.5])
df["Year"]=df["Joining data"].dt.year
df["Month"]=df["Joining data"].dt.month
df["Month name"]=df["Joining data"].dt.month_name()
df["Date"]=df["Joining data"].dt.day_name()
df=df.sort_values(["Date","Month"])
display(df)
```

```python
d=df[df["Month"]==10].index
d
```

```python
dr=df.groupby("Month").size()
plt.pie(dr,labels=dr.index)
plt.show()
```

```python
df=pd.DataFrame({"id":range(1,101)})
df["Joining data"]=pd.to_datetime(np.random.choice(pd.date_range("2020-01-01","2020-12-31"),size=len(df)))
df["Department"]=np.random.choice(("a","f","hr","it"),size=100,p=[0.25,0.25,0.25,0.25])
df["job type"]=np.random.choice(("Part time","full time"),size=100,p=[0.2,0.8])
display(df)
```

```python
dr=df.groupby(["Department","job type"]).size().unstack(fill_value=0)
dr.plot(kind="bar")
```

```python
df["month"]=df["Joining data"].dt.month
dh=df.groupby(["month","Department"]).size().unstack(fill_value=0)

dh.plot(kind="bar")
```

```python
import matplotlib.pyplot as plt

# Add a month column for grouping
df["Month"] = df["Joining data"].dt.to_period("M")

# Count employees per department per month
grouped = df.groupby(["Month", "Department"]).size().unstack(fill_value=0)

# Plot
grouped.plot(kind="bar", figsize=(12,6))
plt.title("Employees Joined per Department by Month")
plt.xlabel("Month")
plt.ylabel("Number of Employees")
plt.xticks(rotation=45)
plt.legend(title="Department")
plt.tight_layout()
plt.show()
```

```python
import random
from datetime import datetime, timedelta

start_date = datetime(2020, 1, 1)
end_date   = datetime(2020, 12, 31)

# Generate random number of days between the range
random_days = random.randint(0, (end_date - start_date).days)

# Add offset to start_date
random_date = start_date + timedelta(days=random_days)

print(random_date)
```
