# Line Plot

## Plot the number of shows released per year as a line plot.

```python
plt.figure(figsize=(10,4))
f=df.groupby("release_year").size()
plt.plot(f,linestyle="--",marker="*",color="red",linewidth=3)
plt.title("Number of shows released per year",color="blue",size=30)
plt.xlabel("Year",color="blue",size=20)
plt.ylabel("Number of shows",color="blue",size=20)
plt.grid()
plt.show()
```

## Show the cumulative number of Netflix releases over time.

```python
df['count'] = 1
cumulative = df.groupby('release_year')['count'].sum().cumsum()

cumulative.plot(marker='o', title="Cumulative Netflix Releases Over Time")
plt.xlabel("Year")
plt.ylabel("Cumulative Count")
plt.show()
```

## Plot the number of shows released per year as a line plot.

```python
df['count'] = 1

releases_per_year = df.groupby('release_year')['count'].sum()
releases_per_year.plot(marker='o', title="Number of Netflix Releases Per Year")
plt.xlabel("Year")
plt.ylabel("Count of Releases")
plt.show()
```

# Bar Plot

## Create a bar plot showing the total number of Movies vs TV Shows.

```python
t=df.groupby("type").size()
plt.bar(t.index,t,color=['blue','green'],edgecolor='black',linewidth=2)
plt.title("Total number of shows",color="blue",size=30)
plt.xlabel("Shows",color="blue",size=20)
plt.ylabel("Number of shows",color="blue",size=20)
plt.grid()
plt.show()
```

## Plot the top 10 countries with the most Netflix titles.

```python
q=df.groupby("country")["title"].value_counts()
plt.bar(q.index,q)
plt.show()
```

## Make a bar plot of the distribution of ratings (TV-MA, PG, R, etc.).

```python
s=df.groupby("rating").size()
plt.bar(s.index,s)
plt.show()
```

## Plot a horizontal bar chart of directors with the most titles (top 10).

```python
k=df.groupby("director")["title"].size()
plt.barh(k.index,k)
plt.show()
```

## Create a bar plot of ratings per content type (Movie/TV Show).

```python
e=df.groupby("rating")["type"].size()
plt.bar(e.index,e)
plt.show()
```

# Pie plot

## Make a pie chart of the distribution of ratings (TV-MA, PG, R, etc.).

```python
er=df.groupby("rating").size()
plt.pie(er,labels=er.index,autopct='%1.2f%%',explode=[0.1,0,0,0,0,0,0])
plt.title("Distribution of Rating",color='indigo',size=20)
plt.show()
```

## Create a pie chart showing the proportion of Movies vs TV Shows.

```python
qw=df.groupby("type").size()
plt.pie(qw,labels=qw.index,autopct='%1.1f%%')
plt.show()
```

## Show a pie chart of the top 5 ratings, grouping the rest as Other.

```python
er=df.groupby("rating").size()
top5=er[:5]
ots=pd.Series({'others':er[5:].sum()})
fin=pd.concat([top5,ots])
plt.pie(fin,labels=fin.index,autopct='%1.2f%%')
plt.show()
```

## Create a pie chart of the top 5 genres.

```python
rt=df.groupby("listed_in").size()
top5=rt[:5]
plt.pie(top5,labels=top5.index,autopct='%1.2f%%')
plt.show()
```

## Compare movies vs TV shows released per year on the same line plot.

```python
plt.figure(figsize=(10,10))
g=df.groupby("release_year").size()
plt.pie(g,labels=g.index,autopct='%1.1f%%')
plt.show()
```

## Plot a pie chart showing the distribution of titles by Type (Movie vs TV Show). Highlight the category with more titles.
