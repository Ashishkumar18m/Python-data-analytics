```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
x1 = [1, 2, 3, 4, 3]
y1 = [2, 4, 2, 8, 10]


plt.figure(figsize=(3, 3))  
plt.plot(x, y, label='Linear', color='blue', linestyle='--', linewidth=2, marker='o')

plt.plot(x1, y1, label='Linear', color='red', linestyle='--', linewidth=2, marker='*')

plt.xlabel('X-axis', fontsize=15)
plt.ylabel('Y-axis', fontsize=15)
plt.title('Basic Line Plot', fontsize=14)


plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=10)


plt.show()
```

```python
c = ['A', 'B', 'C', 'D']
v = [10, 15, 7, 12]

plt.figure(figsize=(3,3))
plt.bar(c, v)
plt.plot(c,v,linestyle="--",color='red',marker='*')
plt.title("Bar Chart")

plt.show()

```

```python

labels = ['Python', 'JavaScript', 'C++', 'Java', 'Other']
sizes = [45, 20, 15, 15,5]
explode = (0, 0, 0.1, 0, 0)  
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'orange']

plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')  
plt.title('Programming Language Popularity', fontsize=14)
plt.show()
```

```python
data = np.random.randn(1000)

plt.figure(figsize=(8, 5))
plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)

plt.title('Normal Distribution', fontsize=14)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D


fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Create data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Plot surface
surf = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')

# Add color bar
fig.colorbar(surf, shrink=0.5, aspect=5)

# Add labels and title
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('3D Surface Plot')

plt.show()
```

```python
z = np.linspace(0, 1, 1002)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, label='3D Line')
ax.set_title('3D Line Plot')
ax.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data
np.random.seed(42)
x = np.arange(10)
y = np.random.rand(10)
df = pd.DataFrame({"Category": ["A", "B", "C", "D"], "Values": [5, 10, 7, 12]})
data_hist = np.random.normal(0, 1, 1000)

# Create a 1x3 subplot grid
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# 1. Seaborn line plot (first subplot)
sns.lineplot(x=x, y=y, ax=axes[1], marker='o', color='green')
axes[0].set_title("Seaborn Line Plot")

# 2. Pandas bar plot (second subplot)
df.plot(kind='bar', x='Category', y='Values', ax=axes[0], color='skyblue')
axes[1].set_title("Pandas Bar Plot")


# 3. NumPy-generated histogram (third subplot)
axes[2].hist(data_hist, bins=30, color='orange', edgecolor='black')
axes[2].set_title("NumPy Histogram")

# Adjust layout and display
plt.tight_layout()
plt.show()
```
