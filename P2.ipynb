# https://www.kaggle.com/datasets/shivamb/netflix-shows?resource=download
## About this Dataset: Netflix is one of the most popular media and video streaming platforms. They have over 8000 movies or tv shows available on their platform, as of mid-2021, they have over 200M Subscribers globally. This tabular dataset consists of listings of all the movies and tv shows available on Netflix, along with details such as - cast, directors, ratings, release year, duration, etc

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

# Read file

```python
df=pd.read_csv("C:\\Users\\ashish\\Downloads\\archive.zip")
display(df.head())
display(df.tail(5))
```

# Find null(nan)

```python
print(df.isnull().sum())
```

# Drop na

```python
# Drop rows with missing show_id or title
df = df.dropna(subset=['show_id', 'title'])
```

```python
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')

# Fill missing date_added with a placeholder
df['date_added'] = df['date_added'].fillna('Not Available')
```

```python
display(df)
```

## Q1: How many shows and movies are in the dataset?

```python
df['type'].value_counts()
```

## Q2: Which genres are most common?

```python
genre_counts = df['listed_in'].str.split(', ').explode().value_counts()
genre_counts.head(10)
```

## Q3. Count of Movies vs TV Shows

```python
sns.countplot(x='type', data=df)
plt.title('Count of Movies vs TV Shows')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

## Q4. Distribution of Movie Durations

```python
df_movies = df[df['type']=='Movie']

plt.figure(figsize=(12,6))
sns.histplot(df_movies['duration'], bins=30, kde=True, color='green')
plt.title('Distribution of Movie Durations (Minutes)')
plt.xlabel('Duration (Minutes)')
plt.ylabel('Number of Movies')
plt.show()
```

## Q5. Ratings Distribution

```python
rating_counts = df['rating'].value_counts()

plt.figure(figsize=(8,8))
plt.pie(rating_counts, labels=rating_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Ratings')
plt.show()
```

## Q6.Horizontal Bar Plot: Top 10 Countries

```python
top_countries = df['country'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_countries.values, y=top_countries.index)
plt.title('Top 10 Countries Producing Content')
plt.xlabel('Number of Shows')
plt.ylabel('Country')
plt.show()
```

## Q7. No. of Movies per year

```python

df_movies = df[df['type'] == 'Movie']

movies_per_year = df_movies['release_year'].value_counts().sort_index()

plt.figure(figsize=(10,8))
sns.barplot(x=movies_per_year.values, y=movies_per_year.index)
plt.title('Number of Movies Released Per Year')
plt.xlabel('Number of Movies')
plt.ylabel('Release Year')
plt.show()
```

## Q8. Top 10 Countries Producing Content

```python
top_countries = df['country'].value_counts().head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_countries.values, y=top_countries.index)
plt.title('Top 10 Countries Producing Content')
plt.xlabel('Number of Shows')
plt.ylabel('Country')
plt.show()
```

## Q9. Movies vs TV Shows per Year (Comparison)

```python
movies_per_year = df[df['type'] == 'Movie'].groupby('release_year').size()
tv_per_year = df[df['type'] == 'Tv Show'].groupby('release_year').size()

plt.figure(figsize=(12,6))
plt.plot(movies_per_year.index, movies_per_year.values, marker='o', label='Movies', color='blue')
plt.plot(tv_per_year.index, tv_per_year.values, marker='s', label='TV Shows', color='orange')
plt.title('Movies vs TV Shows Released Per Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.legend()
plt.grid(True)
plt.show()
```

## Q10. Rolling Average Trend (Smoothed Line)

```python
movies_per_year = df_movies['release_year'].value_counts().sort_index()
rolling_avg = movies_per_year.rolling(window=3).mean()

plt.figure(figsize=(12,6))
plt.plot(movies_per_year.index, movies_per_year.values, label='Actual', color='lightgray', linestyle='--')
plt.plot(rolling_avg.index, rolling_avg.values, label='3-Year Rolling Average', color='blue', linewidth=2)
plt.title('Smoothed Trend: Movies Released per Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Movies')
plt.legend()
plt.grid(True)
plt.show()
```

## Q11. Movies Added vs. Released

```python
plt.figure(figsize=(14,6))
sns.boxplot(x='release_year', y='duration', data=df_movies, showfliers=False)
plt.title('Distribution of Movie Durations by Release Year')
plt.xlabel('Release Year')
plt.ylabel('Duration (minutes)')
plt.xticks(rotation=45)
plt.show()
```

## Q12. Movies per Year by Country (Top 5 Countries Trend)

```python
top5_countries = df[df['type']=='Movie']['country'].value_counts().head(5).index
country_year = df[df['country'].isin(top5_countries) & (df['type']=='Movie')]
country_year = country_year.groupby(['release_year','country']).size().reset_index(name='count')

plt.figure(figsize=(12,6))
sns.lineplot(data=country_year, x='release_year', y='count', hue='country', marker='o')
plt.title('Movies Released per Year by Top 5 Countries')
plt.xlabel('Release Year')
plt.ylabel('Number of Movies')
plt.legend(title='Country')
plt.grid(True)
plt.show()
```

## Q13. Grouped Bar Plot â€“ Movies vs TV Shows per Year

```python
movies_per_year = df[df['type']=='Movie'].groupby('release_year').size().reset_index(name='Movies')
tv_per_year = df[df['type']=='Tv Show'].groupby('release_year').size().reset_index(name='TV Shows')

combined = pd.merge(movies_per_year, tv_per_year, on='release_year', how='outer').fillna(0)
combined_melt = combined.melt(id_vars='release_year', var_name='Type', value_name='Count')

plt.figure(figsize=(12,6))
sns.barplot(data=combined_melt, x='release_year', y='Count', hue='Type')
plt.title('Movies vs TV Shows Released per Year')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

## Q14. Scatter Plot â€“ Release Year vs Duration

```python
df_movies = df[df['type']=='Movie']
plt.figure(figsize=(12,6))
sns.scatterplot(x='release_year', y='duration', data=df_movies, alpha=0.5, color='navy')
sns.scatterplot(x='release_year', y='duration', data=df_movies, scatter=False, color='red')
plt.title('Movie Duration vs Release Year')
plt.xlabel('Release Year')
plt.ylabel('Duration (minutes)')
plt.show()
```
