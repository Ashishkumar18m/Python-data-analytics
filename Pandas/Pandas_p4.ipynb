# Reading a file

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("C:\\Users\\ashish\\Downloads\\sample_employee_data.csv")
display(df.head(4))
```

## Q find the avg salary for each dept an make bar chart

```python
avg_salary = df.groupby('Department') ['Salary'].mean()
plt.figure(figsize=(10,4))
avg_salary.plot(kind='bar', color=['orange', 'red', 'green','blue','yellow'],edgecolor='black',       
    linewidth=1.5)

plt.title("Average Salary by Department", color='blue',fontsize=20, fontweight='bold')
plt.ylabel("Average Salary ($) " ,color='brown', fontsize=14,fontweight='bold')
plt.xlabel("Department" ,color='red', fontsize=14,fontweight='bold')

plt.show()
```

## count how many employees work remotly and display pie chart

```python
d_counts = df['Remote_Work'].value_counts()
plt.figure(figsize=(4, 4))
explode=(0,0.1)
plt.pie(d_counts, labels=d_counts.index, autopct='%1.1f%%',textprops={'color': 'white'},explode=explode,shadow=True )
plt.title('Employees per Department',fontsize=20,color='green')
plt.show()
```

 ## Q plot a scatter plot showing the relation ship b/w age a snd salary

```python
plt.figure(figsize=(10,4))
#avg_salary.plot(kind='bar', color=['orange', 'red', 'green','blue','yellow'],edgecolor='black',       
 #   linewidth=1.5)
plt.scatter(df['Age'],df['Salary'],color='red')
plt.title("Relation b/w Age & Salary", color='blue',fontsize=20, fontweight='bold')
plt.ylabel("Salary " ,color='brown', fontsize=14,fontweight='bold')
plt.xlabel("Age" ,color='red', fontsize=14,fontweight='bold')
plt.grid(
    alpha=1,color='black',linestyle='--'
)
plt.show()
```

## Q find th total bonus paid per dept make horizontal bar chart 

```python
avg_salary = df.groupby('Department') ['Bonus'].mean()
plt.figure(figsize=(10,4))
avg_salary.plot(kind='barh', color=['orange', 'red', 'green','blue','yellow'],edgecolor='black',       
    linewidth=1.5)

plt.title("Total bonus paid per Department", color='blue',fontsize=20, fontweight='bold')
plt.ylabel("Departement " ,color='brown', fontsize=14,fontweight='bold')
plt.xlabel("Bonus" ,color='red', fontsize=14,fontweight='bold')

plt.show()
```

## Q create a line plot showing trend of salary against joining date 

```python
df['JoiningDate'] = pd.to_datetime(df['Joining_Date'])

sorted_df = df.sort_values('Joining_Date')

plt.figure(figsize=(12, 5))
plt.plot(sorted_df['Joining_Date'], sorted_df['Salary'], color='red',linewidth = 3, linestyle = '--')
plt.title('Salary Trend by Joining Date')
plt.xlabel('Joining Date')
plt.ylabel('Salary')
plt.xticks(rotation=45, fontsize=8)
plt.tight_layout()
plt.show()
```
