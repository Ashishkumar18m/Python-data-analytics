```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(8)
sale_ids = np.arange(1, 9)
products = np.random.choice(["Laptop", "Phone", "Tablet"], size=8)
amounts = np.random.randint(5000, 50001, size=8)


df_sales = pd.DataFrame({
    "SaleID": sale_ids,
    "Product": products,
    "Amount": amounts
})

s=np.random.randint(1,9,8)
n = ["Alice", "Bob", "Charlie", "Diana", "Ethan", "Fiona", "George", "Hannah"]
df_customer=pd.DataFrame({
    "SaleID":s,
    "Customer_name":n
    })
e=pd.merge(df_sales,df_customer,on="SaleID",how="inner")
print(e)
```

```python
e=pd.merge(df_sales,df_customer,on="SaleID",how="inner")
print(e)
f=pd.merge(df_sales,df_customer,on="SaleID",how="outer")
print(f)
g=pd.merge(df_sales,df_customer,on="SaleID",how="left")
print(g)
u=pd.merge(df_sales,df_customer,on="SaleID",how="right")
print(u)
```

```python

threshold = np.random.randint(20000, 40001)
print(threshold)


filtered_sales = df_sales[df_sales["Amount"] > threshold]
plt.figure(figsize=(4, 4))

plt.bar(filtered_sales["Product"], filtered_sales["Amount"])

plt.xlabel("Product")
plt.ylabel("Sales Amount")
plt.title(f"Products with Amount > {threshold}")
plt.show()
```
